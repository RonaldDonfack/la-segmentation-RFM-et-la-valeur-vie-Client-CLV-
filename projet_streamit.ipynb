```python
1. Fiche synthétique du dataset
Le fichier utilisé est le document Excel « online_retail_II.xlsx », dérivé du jeu de données 
Online Retail II du UCI Machine Learning Repository. Il contient les transactions e-commerce 
d’un détaillant basé au Royaume-Uni.

Le fichier est structuré en deux onglets distincts :
* « Year 2009-2010 »
* « Year 2010-2011 »

Chaque onglet regroupe l’ensemble des ventes réalisées durant sa période respective.
L’unité d’observation est la ligne de transaction, c’est-à-dire un produit acheté dans une facture donnée.
Période couverte :
* Entre le 1er décembre 2009 et le 9 décembre 2011.
Granularité :
* Une ligne = un produit sur une facture.
* Niveau d’analyse principal : transaction, avec possibilité d’agrégation par facture, par client, ou par cohorte.

Colonnes principales :
* Invoice : numéro de facture.
* StockCode : identifiant du produit.
* Description : nom du produit.
* Quantity : quantité achetée (valeurs négatives possibles en cas de retour).
* InvoiceDate : date et heure de la transaction.
* Price : prix unitaire.
* Customer ID : identifiant du client.
* Country : pays du client.
    
Ces données serviront à construire les cohortes d’acquisition, la segmentation RFM, 
le calcul de la valeur vie client (CLV), ainsi que l’évolution du revenu par cohorte.
```

```python
Voici une description simple et rapide des variables du fichier.

Invoice
Numéro de facture. Si le code commence par « C », c’est une facture annulée/avoir. 
Sert surtout à identifier les transactions et gérer les retours.

StockCode
Identifiant du produit. 
Utile pour repérer les produits, analyser les paniers et faire des regroupements.

Description
Nom du produit. 
Variable surtout informative.

Quantity
Quantité achetée. 
Valeurs négatives = retours. 
Sert au calcul du volume et du revenu.

InvoiceDate
Date et heure de la transaction. 
Variable clé pour les cohortes, l’analyse temporelle et la récence RFM.

Price
Prix unitaire. 
Permet de calculer le montant total d’une transaction (Quantity × Price).

Customer ID
Identifiant du client. 
Souvent manquant. Indispensable pour les cohortes, RFM et CLV.

Country
Pays du client. 
Sert à segmenter par zone géographique.
```

```python
Analyse de la qualité des données
3. Qualité des données

L’analyse de la qualité des données est essentielle, car la construction des cohortes, du RFM 
et de la CLV dépend fortement de la précision des informations.

a. Valeurs manquantes

La colonne « Customer ID » contient un volume important de valeurs manquantes. Ces transactions ne permettent 
pas une analyse par client et devront être exclues des parties RFM, cohortes et CLV.
La colonne « Description » présente aussi quelques absences, sans impact majeur sur les analyses quantitatives.

b. Doublons

Il est nécessaire d’identifier et de supprimer les doublons exacts, ainsi que les doublons potentiels basés 
sur les colonnes suivantes :
Invoice, StockCode, Quantity, InvoiceDate, Customer ID.

c. Outliers

Les quantités peuvent présenter des valeurs extrêmes, parfois anormales, ainsi que des valeurs négatives 
indiquant des retours.
Le prix unitaire peut être égal à zéro ou exceptionnellement élevé ; ces cas devront être vérifiés.

d. Gestion des factures annulées

Les factures commençant par la lettre « C » représentent des annulations.
Elles doivent être traitées à part :

Exclues de l’analyse d’acquisition (une annulation n’est pas un nouvel achat).

Intégrées dans le calcul du revenu net (pour prendre en compte les retours).

e. Granularité temporelle

Les transactions sont horodatées à la minute, ce qui permet une analyse fine mais nécessite 
une agrégation mensuelle pour la construction des cohortes et des métriques RFM.
Une saisonnalité forte est attendue, en particulier en fin d’année.
```
